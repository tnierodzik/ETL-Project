```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

### E&T for Blue Jays Schedule

```python
csv_file = "Resources/toronto_blue_jays_home_game_scheduled_start_and_finish_times_2008_to_2018.csv"
schedule_df = pd.read_csv(csv_file)
# schedule_df.info()
schedule_df.head()
```

```python
schedule_df['game_start'] = pd.to_datetime(schedule_df['game_start'])
schedule_df['Date'] = schedule_df['game_start'].dt.date
schedule_df
```

```python
schedule_df2 = schedule_df[['Date',
                               'team',
                               'opponent',
                               'home_score',
                               'away_score',
                               'game_duration',
                               'attendance'
                               ]].copy()
```

```python
#Renaming the schedule columns
schedule_df3 = schedule_df2.rename(columns = {
                               'team':'Home Team',
                               'opponent':'Away Team',
                               'home_score': 'Home Score',
                               'away_score': 'Away Score',
                               'game_duration':'Game Duration',
                               'attendance': 'Attendance'})
schedule_df3.head()
```

### E&T for Toronto Demographics

```python
#reading in the demographics data
csv_file = "Resources/toronto_demographic_data.csv"
demo_df = pd.read_csv(csv_file)
demo_df2 = demo_df[demo_df['Year'] > 2007]
demo_df3 = demo_df2[demo_df2['Year'] < 2019]
```

```python
table = pd.pivot_table(demo_df3, values='Value', index=['Year','Period'], columns=['Measure Name'], aggfunc=np.sum).reset_index()
table.head(100)
```

```python
table2 = table.rename(columns = {
    "Period" : "Month",
    "CPI - Consumer Price Index - Toronto (% Change)" : "Consumer Price Index (% Change)",
    "Crime Total Number of Incidents (in 7 Crime Categories)" : "Total Number of Crime Incidents",
    "Unemployment Rate (Toronto Residents)" : "Unemployement Rate"
})
table2
```
